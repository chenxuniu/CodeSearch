```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#download the dataset from the UCI machine learning repository and put it into a data folder in the working directory
data = pd.read_csv("./data/SMSSpamCollection",delimiter='\t',header=None,names=["classification","text"])
data['target'] = np.where(data['classification']=='ham',0, 1)
data['target2'] = np.where(data['classification']=='ham',1, 0)
print(data.head())
data.shape
```

```python
#clean text, stem or lemmatize, remove stop words
def text_clean2(text, method, rm_stop):
    text = text.replace("</p>","") # removing </p>
    text = text.replace("<p>"," ")  # removing <p>
    text = text.replace("http", " ")
    text = text.replace("www", " ")
    text = re.sub(r'([a-z])\1+', r'\1', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub('\.+', '.', text)
    text = re.sub(r"(?:\@|'|https?\://)\s+","",text) #delete punctuation
    text = re.sub("[^a-zA-Z]", " ",text)
    text = re.sub(r"\d+","",text)   #remove numbers with one or more digits
    text = re.sub(r'[^\x00-\x7f]',r' ',text)   #remove non-ascii (everything without a code from x00 to x7f)
    text = re.sub(r'[^\w\s]','',text)   #remove punctuation (everything not a word or space)
    text = re.sub(r"\n","",text)   #remove line breaks
    text = text.lower() #convert to lowercase
    
    #remove stop words 
    if rm_stop == True:
        filtered_tokens = [word for word in word_tokenize(text) if not word in set(stopwords.words('english'))]
        text = " ".join(filtered_tokens)

    #lemmatization: typically preferred over stemming
    if method == 'L':
        lemmer = WordNetLemmatizer()
        lemm_tokens = [lemmer.lemmatize(word) for word in word_tokenize(text)]
        text = " ".join(lemm_tokens)
    
    #stemming 
    if method == 'S':
        porter = PorterStemmer()
        stem_tokens = [porter.stem(word) for word in word_tokenize(text)]
        text = " ".join(stem_tokens)
    
    #delete tokens of size 2 or less
    trunc_tokens=[w for w in word_tokenize(text) if len(w)>2] # delete tokens of size 2 or less
    text = " ".join(trunc_tokens)
    
    return text
```

```python
#clean the text
sentences = data.text
preprocessed_sentences = [text_clean2(text, 'L', True) for text in sentences]
len(preprocessed_sentences)
```

```python
from collections import defaultdict
#default value returned for a defaultdict
def def_value():
    return 0

#collect the ham and spam tokens separately
doc_ham = defaultdict(def_value)
doc_spam = defaultdict(def_value)
i = 0
for text in preprocessed_sentences:
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if data.iloc[i].target:
            doc_spam[w] += 1
        else:
            doc_ham[w] += 1
    i += 1
```

```python
#ham word frequency
print(sorted(doc_ham.items(), key=lambda kv: 
                 (kv[1], kv[0])))
```

```python
#spam word frequency
print(sorted(doc_spam.items(), key=lambda kv: 
                 (kv[1], kv[0])))
```

```python
print('Number of ham words:',len(doc_ham.keys()))
print('Number of spam words:',len(doc_spam.keys()))
```

```python
#find words in common between ham and spam
com_words = []
i = 0
for w in doc_spam.keys():
    if w in doc_ham.keys():
        i += 1
        #print(w,'\t\t\t',doc_ham[w],'\t\t\t',doc_spam[w])
        com_words.append(w)
print('Number of words in common:',i)
print('Number of words in common:',len(com_words))
```

```python
#option to try or leave out
#delete tokens common to ham and spam
def remove_common_tokens (text, comwords):
    diff_tokens=[w for w in word_tokenize(text) if w not in comwords] # delete common tokens
    text = " ".join(diff_tokens)
    return text
```

```python
preprocessed_sentences = [remove_common_tokens(text, com_words) for text in preprocessed_sentences]
```

```python
#option to try or leave out
#remove words with frequencies lower than num
def remove_words (text, dt, num):
    filtered_tokens = [word for word in word_tokenize(text) if dt[word] > num]
    text = " ".join(filtered_tokens)
    return text
```

```python
rm_list = []
i = 0
for text in preprocessed_sentences:
    if data.iloc[i].target:
        rm_list.append(remove_words(text, doc_spam, 4))
    else:
        rm_list.append(remove_words(text, doc_ham, 4))
    i += 1
preprocessed_sentences = rm_list
len(preprocessed_sentences)
```

```python
#deterimine how many sentences are now empty
i = 0
j = 0
spm = 0
for s in preprocessed_sentences:
    if len(s) == 0:
        i += 1
        if data.iloc[j].target:
            spm += 1
    j += 1
print ("Num of empty preprocessed sentences:",i)
print("Num of empty spam sentences",spm)
```

```python
vectorizer3 = TfidfVectorizer() 
vectorized_documents3 = vectorizer3.fit_transform(preprocessed_sentences)
vectorized_documents3
```

```python
# reduce the dimensionality of the data using PCA for graphing the data
pca3 = PCA(n_components=2) 
reduced_data3 = pca3.fit_transform(vectorized_documents3.toarray())
reduced_data3
```

```python
# cluster the documents using k-means 
num_clusters = 2
kmeans3 = KMeans(n_clusters=num_clusters, n_init=5, 
                max_iter=500, random_state=42) 
kmeans3.fit(vectorized_documents3)
```

```python
# create a dataframe to store the results 
results3 = pd.DataFrame() 
results3['document'] = preprocessed_sentences
results3['cluster'] = kmeans3.labels_
```

```python
#plot the true ham emails
colors = ['blue', 'orange'] 
cluster = ['Ham','Spam']
plt.scatter(reduced_data3[data.target == 0, 0], 
                reduced_data3[data.target == 0, 1],  
                s=10, color=colors[0],  
                label=f' {cluster[0]}')
plt.legend() 
plt.show()
```

```python
#plot the true spam emails
colors = ['blue', 'orange'] 
cluster = ['Ham','Spam']
plt.scatter(reduced_data3[data.target == 1, 0], 
                reduced_data3[data.target == 1, 1],  
                s=10, color=colors[1],  
                label=f' {cluster[1]}')
plt.legend() 
plt.show()
```

```python
cm = confusion_matrix(data.target,results3.cluster)
print(cm)
print ("Accuracy score:", accuracy_score(data.target, results3.cluster))
```
